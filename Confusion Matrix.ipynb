```python
import numpy as np
actual = np.random.binomial(1,0.9,1000)
predicted = np.random.binomial(1,0.9,1000)

from sklearn import metrics
mat=metrics.confusion_matrix(actual,predicted)
mat_display=metrics.ConfusionMatrixDisplay(confusion_matrix=mat,display_labels=[False,True])

mat_display.plot()
```

```python
print("Accuracy:",metrics.accuracy_score(actual,predicted))
print(":",metrics.accuracy_score(actual,predicted))
```
