# Polynomial Regression

```python
# The x-axis represents the hours of the day and the y-axis represents the speed
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
import numpy as np

# It's important to choose an appropriate degree for the polynomial 
# to avoid overfitting or underfitting the data.
model=np.poly1d(np.polyfit(x,y,3))
line=np.linspace(1,22,100) #creating line of 100 parts between 1 and 22
plt.scatter(x,y)
plt.plot(line,model(line),color='y')
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print("Score:",r2_score(y,model(x))*100,'%')
```

```python
# also other metrics are:
print("Mean Absolute Error:",mean_absolute_error(y,model(x)))
print("Mean Squared Error:",mean_squared_error(y,model(x)))
```

# Prediction

```python
# Let's predict speed from hour
hour=17
speed=model(hour)
print("speed:",speed)
```

```python
plt.scatter(x,y)
plt.plot(line,model(line),color='y')
plt.scatter(hour,speed,color='r')
plt.show()
```

# Bad Fit?

```python
x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

model=np.poly1d(np.polyfit(x,y,3))
line=np.linspace(2,95,100)

plt.scatter(x,y)
plt.plot(line,model(line),color='y')
```

```python
print("Score:",r2_score(y,model(x))*100,'%')
```

```python
print("Mean Absolute Error:",mean_absolute_error(y,model(x)))
print("Mean Squared Error:",mean_squared_error(y,model(x)))
```
