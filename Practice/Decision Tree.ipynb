```python
# Reading csv
import pandas as pd
df=pd.read_csv('../Datasets/dt.csv')
df
```

```python
# Changing categorical values to numerical values
d={'UK':0,'USA':1,'N':2}
df['Nationality']=df['Nationality'].map(d)
d={'NO':0,'YES':1}
df['Go']=df['Go'].map(d)
df
```

```python
# Splittng data
features=['Age','Experience','Rank','Nationality']
x=df[features]
x
```

```python
y=df['Go']
y
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x,y)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
tree.plot_tree(model,feature_names=features)
```

# Prediction

```python
print("Go:",model.predict([[1,1,6.49,1]]))
print("Go:",model.predict([[1,1,6.5,1]]))
print("Go:",model.predict([[1,0,6.51,1]]))
print("Go:",model.predict([[1,9.5,6.51,1]]))
print("Go:",model.predict([[1,9.51,6.51,1]]))
print("Go:",model.predict([[1,11.51,6.51,1]]))
```
