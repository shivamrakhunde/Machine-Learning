```python
import numpy as np
np.random.seed(2)
x=np.random.normal(3,1,100)
y=np.random.normal(150,40,100)/x

import matplotlib.pyplot as plt
plt.scatter(x,y)
```

# Split into train test

```python
xtrain=x[:80]
ytrain=y[:80]
xtest=x[80:]
ytest=y[80:]
```

```python
# Display the training set
plt.scatter(xtrain,ytrain)
```

```python
# Display the testing set
plt.scatter(xtest,ytest)
```

# Fitting the data set

```python
model=np.poly1d(np.polyfit(xtrain,ytrain,4))
line=np.linspace(0,6,100)

plt.scatter(x,y)
plt.plot(line,model(line),color='y')
```

```python
from sklearn.metrics import r2_score
print("score:",r2_score(ytrain,model(xtrain))*100,'%')
```

# Bring in the Testing Set

```python
print("score:",r2_score(ytest,model(xtest))*100,'%')
```

# Prediction

```python
a=5
b=model(5)
plt.scatter(x,y)
plt.plot(line,model(line),color='y')
plt.scatter(a,b,color='r')
```
