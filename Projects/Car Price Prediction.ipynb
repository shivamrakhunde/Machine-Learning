```python
# Encoding the Categorical Data
from sklearn.preprocessing import LabelEncoder
lbl=LabelEncoder()
df=df.apply(lambda x: lbl.fit_transform(x) if x.dtype=='object' else x)
df
```

```python
# Correlation Matrix (relationship between various columns)
df.corr()
```

```python
# Visualize the Data
import seaborn as sns
sns.heatmap(df.corr(),annot=True)
```

```python
# Dependent Variable
dependent=['Year','Present_Price','Kms_Driven','Fuel_Type','Seller_Type','Transmission','Owner']
x=df[dependent]

# OR

x=df.drop(['Car_Name','Selling_Price'],axis=1)
x
```

```python
y=df['Selling_Price']
y
```

# Linear Regression

```python
# Splitting the dataset into training and testing data
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
# Model Training
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
# Prediction on training data
ytrainpred=model.predict(xtrain)
```

```python
from sklearn import metrics
train_error=metrics.r2_score(ytrain,ytrainpred)
print("Score:",train_error*100,'%')
```

```python
import matplotlib.pyplot as plt
plt.scatter(ytrain,ytrainpred)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```

```python
# Prediction on testing data
ytestpred=model.predict(xtest)
test_error=metrics.r2_score(ytest,ytestpred)
print("Score:",test_error*100,'%')
```

```python
plt.scatter(ytest,ytestpred)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```
